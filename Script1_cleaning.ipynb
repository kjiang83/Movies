```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # adjusts config of plots
```

```python
# Read in data
df = pd.read_csv(r"C:\Users\Kevin Jiang\Documents\GitHub\Movies\movies.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df_filled = df.copy()
```

```python
for col in df_filled.columns:
    pct_missing = np.mean(df_filled[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df_filled["rating"].fillna("Unavailable")
df_filled["released"].fillna("Unavailable")
df_filled["score"].fillna("Unavailable")
df_filled["writer"].fillna("Unavailable")
df_filled["star"].fillna("Unavailable")
df_filled["country"].fillna("Unavailable")
df_filled["company"].fillna("Unavailable")
```

```python
df_filled['votes'] = pd.to_numeric(df_filled['votes'], errors = 'coerce').fillna(0).astype(np.int64)
df_filled['budget'] = pd.to_numeric(df_filled['budget'], errors = 'coerce').fillna(0).astype(np.int64)
df_filled['gross'] = pd.to_numeric(df_filled['gross'], errors = 'coerce').fillna(0).astype(np.int64)
```

```python
df_filled.dtypes
```

```python
df_coryear = df_filled.copy()
```

```python
df_coryear.head()
```

```python
# split 'released' column and get correct year
df_coryear['yearcorrect'] = df_coryear['released'].str.split(expand = True)[2]
```

```python
df_coryear.head()
```

```python
df_coryear.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
df_dupes = df_coryear.copy()
```

```python
# drop duplicates
df_dupes.drop_duplicates()
```

```python
df_clean = df_dupes.copy()
```

```python
# budget vs gross
plt.scatter(x = df_clean['budget'], y = df_clean['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
df_clean.head().sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
sns.regplot(x = 'budget', y = 'gross', data = df_clean, scatter_kws = {'color' : 'mediumorchid'}, line_kws = {'color' : 'indigo'})
```

```python
# pearson correlation coef
df_clean.corr()
```

```python
corr_matrix = df_clean.corr(method = 'pearson')
```

```python
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix of Movie Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_clean.head()
```

```python
df_numerize = df_clean.copy()
```

```python
for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype == 'object'):
        df_numerize[col_name] = df_numerize[col_name].astype('category')
        df_numerize[col_name] = df_numerize[col_name].cat.codes
df_numerize.head()
```

```python
corr_matrix = df_numerize.corr(method = 'pearson')
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix of Movie Variables')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerize.corr()
```

```python
corr_mat = df_numerize.corr()
corr_pairs = corr_mat.unstack()
corr_pairs
```
